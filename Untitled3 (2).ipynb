```python
def minimum(x, y):
    if x <= y:
        return x
    else:
        return y

def maximum(x, y):
    if x >= y:
        return x
    else:
        return y

# Example usage
a = 7
b = 3
print("Minimum:", minimum(a, b))  # Output: 3
print("Maximum:", maximum(a, b))  # Output: 7
```
